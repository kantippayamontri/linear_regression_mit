```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
# load dataset
data = pd.DataFrame(pd.read_csv("./Housing.csv"))
data = data[["area", "price"]]
data['price'] = data["price"] / 10000
# data = data.head(100)
# print(data.head(), data.tail())
```

```python
x = data["area"].values.reshape(-1,1) # numpy array
y = data["price"].values.reshape(-1,1) # numpy array


model = LinearRegression()
model = model.fit(x,y)
```

```python
r_sq = model.score(x,y)
print(f"r square : {r_sq}")
slope = model.coef_[0][0]
intercept = model.intercept_[0]
print(f"slope : {slope}, intercept : {intercept}")


fig, ax = plt.subplots(1, 1, figsize=(60, 30))
# fig.suptitle(f"Round {round+1}, Error = {loss_fn}")

for epochs in range(1):
    theta0 = intercept
    theta1 = slope
    print(f"theta0 = {theta0}, theta1 = {theta1}")
    ax.scatter(data["area"], data["price"])
    ax.plot(
        data["area"],
        [
            ((inp * theta1) + theta0)
            for inp in data["area"]
        ],
    )
plt.show()
```
