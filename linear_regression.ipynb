```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load dataset
data = pd.DataFrame(pd.read_csv("./Housing.csv"))
data.shape[0]
```

```python
data.head()
```

```python
data = data[["area", "price"]]
print(data.head(20))
data['price'] = data["price"] / 1000 
data.head(20)
```

```python
plt.scatter(data["area"], data["price"])
plt.xlabel("area in feet^2")
plt.ylabel("price in dollars")
```

```python
from icecream import ic
def linear_regression_model(inp, theta0, theta1):
    output = theta0 + (theta1 * inp )
    return output

def loss_square_function(actual, predict):
    loss = (predict - actual) ** 2
    return loss

def loss_diff_function(actual, predict):
    loss = predict - actual
    return loss

def gradient_theta0(loss, data_size):
    gradient = (1.0 / data_size) * loss
    return gradient

def gradient_theta1(loss_input, data_size, ):
    gradient = (1.0 / data_size) * loss_input
    return gradient
```

```python
theta0 = 50
theta1 = 0.1

learning_rate = 0.000000001
epochs = 100

theta_his = {}
error_his = []
for e in range(epochs):
    theta_his[e] = {"theta0": theta0, "theta1": theta1}

    sum_square_loss = 0
    sum_diff_loss = 0
    sum_diff_input_loss = 0
    m = data.shape[0]
    for index in range(m):
        inp = data.iloc[index]["area"]
        op = data.iloc[index]["price"]
        predict_value = linear_regression_model(inp=inp, theta0=theta0, theta1=theta1)
        loss_square = loss_square_function(actual=op, predict=predict_value)
        loss_diff = loss_diff_function(actual=op, predict=predict_value)

        sum_square_loss += loss_square
        sum_diff_loss += loss_diff
        sum_diff_input_loss += loss_diff * inp
    loss_fn = (1 / (2 * m)) * (sum_square_loss)
    error_his.append((loss_fn, e))
    theta_his[e]["error"] = sum_square_loss
    print(f"epochs {e+1}, loss function = {loss_fn}")

    theta0 = theta0 - (
        learning_rate * gradient_theta0(loss=sum_diff_loss, data_size=m)
    )
    theta1 = theta1 - (
        learning_rate
        * gradient_theta1(loss_input=sum_diff_input_loss, data_size=m,)
    )
```

```python
number_plot = 10

SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=MEDIUM_SIZE)
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=15)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=15)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

fig, ax = plt.subplots(number_plot, 1, figsize=(20, 50))

for img_num in range(number_plot):
    theta_his_index = int((epochs / number_plot) * img_num)
    theta0 = theta_his[theta_his_index]["theta0"]
    theta1 = theta_his[theta_his_index]["theta1"]
    print(f"round:{int(theta_his_index)}, theta0 = {theta0}, theta1 = {theta1}")
    ax[img_num].scatter(data["area"], data["price"])
    ax[img_num].plot(
        data["area"],
        [
            linear_regression_model(inp=inp, theta0=theta0, theta1=theta1)
            for inp in data["area"]
        ],
        color="red",
        label=f"round {int(theta_his_index)}"
    )
    if theta_his_index > 0:
        prev_index = (epochs / number_plot) * (img_num - 1)
        prev_theta0 = theta_his[prev_index]["theta0"]
        prev_theta1 = theta_his[prev_index]["theta1"]
        ax[img_num].plot(
            data["area"],
            [
                linear_regression_model(inp=inp, theta0=prev_theta0, theta1=prev_theta1)
                for inp in data["area"]
            ],
            color="orange",
            linestyle="dotted",
            label = f"round {int(prev_index)}"
        )

    ax[img_num].legend(loc="upper left",)
    ax[img_num].set_title(
        f"round: {int(theta_his_index)}, error = {theta_his[theta_his_index]['error']},"
    )
    ax[img_num].set_xlabel("area")
    ax[img_num].set_ylabel("price")
    fig.tight_layout(pad=1.0)
plt.show()
```

```python
# Predict program
con = True
best_index = min(error_his)[1]
best_theta0 = theta_his[best_index]["theta0"]
best_theta1 = theta_his[best_index]["theta1"]
# print(f"best index: {best_index}")
while con:
    area = float(input("Please enter the area of the house: "))
    price = linear_regression_model(inp=area, theta0=best_theta0, theta1=best_theta1)
    print(f"the price from model is {price:.2f} x 1000 dollars.")

    # plot the answer
    plt.plot(
        data["area"],
        [
            linear_regression_model(inp=inp, theta0=best_theta0, theta1=best_theta1)
            for inp in data["area"]
        ],
        color="red",
        label = f"prediction line"
    )
    plt.plot(
        [area],
        [price],
        color="green",
        marker="o",
        label=f"price at area={area} feet^2"
        
    )

    con = False
    plt.legend(loc="upper left")
    plt.xlabel("area")
    plt.ylabel("price")
    plt.show()
    # if "-1" == input("Please enter to continue and -1 to exit"):
    # con = False
```
